```python
import matplotlib.pyplot as plt

# Creating a 2x2 grid of subplots


plt.subplot(2, 2, 1)  # First subplot in a 2x2 grid
plt.plot([1, 2, 3])

plt.subplot(2, 2, 2)  # Second subplot in a 2x2 grid
plt.plot([4, 5, 6])

plt.subplot(2, 2, 3)  # Third subplot in a 2x2 grid
plt.plot([7, 8, 9])

plt.subplot(2, 2, 4)  # Fourth subplot in a 2x2 grid
plt.plot([10, 11, 12])

plt.show()
```

```python
import matplotlib.pyplot as plt

# Equivalent to the above example using the compact form
plt.subplot(221)
plt.plot([1, 2, 3])
plt.title("Plot 1")

plt.subplot(222)
plt.plot([4, 5, 6])
plt.title("Plot 2")

plt.subplot(223)
plt.plot([7, 8, 9])
plt.title("Plot 3")

plt.subplot(224)
plt.plot([10, 11, 12])
plt.title("Plot 4")

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Correct: Create subplots using subplots()
fig, ax = plt.subplots(2, 2)  # 2x2 grid of subplots

# Plot on each Axes object
ax[0, 0].plot([1, 2, 3])
ax[0, 1].plot([4, 5, 6])
ax[1, 0].plot([7, 8, 9])
ax[1, 1].plot([10, 11, 12])

plt.show()

```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 14, 19, 25, 30]

# Create a scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Basic Scatter Plot")

# Show the plot
plt.show()
```
