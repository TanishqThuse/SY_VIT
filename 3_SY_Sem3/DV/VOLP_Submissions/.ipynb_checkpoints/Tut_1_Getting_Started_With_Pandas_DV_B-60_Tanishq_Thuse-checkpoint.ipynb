```python
#A python function to assign DV Subject grade
def grade_dv(marks):
    marks = int(marks)
    if marks > 95:
        return 'A+'
    elif marks > 90 :
        return 'A'
    else:
        return 'B'

# Apply the grading function to the DV marks
df['DV Grade'] = df['DV'].apply(grade_dv)

# Display the updated DataFrame
print(df)
```

```python
colours=['Red','Blue','Green']
capacities=[100,200,300]
```

append adds to the list (at last)

```python
capacities.append(400)
```

```python
print(capacities)
```

```python
colours.append('Yellow')
```

```python
print(colours)
```

insert() takes 2 input values first => position on which value is to be inserted, second => value to be inserted

```python
capacities.insert(1,200)
print(capacities)
```

what you want to remove

```python
colours.remove('Red')
```

```python
print(colours)
```

gives the postion of the input value

```python
capacities.index(400)
```

if there are 2 same values then it returns the position of the first one

```python
capacities.index(200)
```

Copies the content of one list to another

```python
a=colours.copy()
print(a)
```

Clears the given list completely

```python
a.clear()
print(a)
```

Counts how many times the given input is repeated in the list
